```python
#Exercício 1, aula 9 - Tuplas

def exercicio_1(n):
    n = 8 # n representa o valor máximo para a tupla de números e seus quadrados
    lista = []

    while n > 0:
        lista.append((n, n**2))
        n -= 1
```

```python
#Exercício 2, aula 9 - Tuplas

def exercicio_2(n):
    n = 8 # n representa o valor máximo para a tupla de números, seus quadrados e seus cubos
    lista = []

    while n > 0:
        lista.append((n, n**2, n**3))
        n -= 1
```

```python
#Exercício 8, aula 9 - Tuplas

def evercicio_7(lista_de_cotacoes):
    lista_de_cotacoes = float(input("Digite a cotação atual do dolar: "))

    #Dólar para Real   
    real = list(range(0, 101))
    dolar_p_real = []
    tupla = ()

    print ("\nUS$ para R$:")
    print ("-----------------------")

    for i in real:
        tupla = (i, i/lista_de_cotacoes)
        dolar_p_real.append(tupla) # relação de dolár para real

    tam = len(dolar_p_real)
    n = 0 #índie de posição dos dados na lista

    while tam > 0:
        print("\nUS$", dolar_p_real[n][0], " para R$:", dolar_p_real[n][1])
        print ("-----------------------")
        tam -= 1
        n += 1

    #Real para Dólar
    dolar = list(range(0, 101))
    real_p_dolar = []
    tupla2 = ()

    print ("\n")
    print ("\nR$ para US$:")
    print ("-----------------------")

    for i in dolar:
        tupla2 = (i, i*lista_de_cotacoes)
        real_p_dolar.append(tupla2) # relação de real para dólar

    tam2 = len(real_p_dolar)
    n = 0 #índie de posição dos dados na lista

    while tam2 > 0:
        print("\nR$", real_p_dolar[n][0], " para US$:", real_p_dolar[n][1])
        print ("-----------------------")
        tam2 -= 1
        n += 1
```
